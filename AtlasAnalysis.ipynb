```python
import pandas as pd

Crime = pd.read_csv('/Users/harryfranklin/Downloads/Crimes.csv')

Crime.head()

Battery = Crime[Crime['Primary Type'] == 'BATTERY']
Assault = Crime[Crime['Primary Type'] == 'ASSAULT']
SA = Crime[Crime['Primary Type'].isin(["CRIMINAL SEXUAL ASSAULT", "CRIM SEXUAL ASSAULT"])]
Homicide = Crime[Crime['Primary Type'] == 'HOMICIDE']

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", None)

Crime.head(5)
```

```python
import geopandas as gpd
from shapely.geometry import Point

tracts = gpd.read_file("/Users/harryfranklin/Downloads/Census_20Tracts/Census_Tracts.shp")

geometry = [
    Point(xy) for xy in zip(Crime["Longitude"], Crime["Latitude"])
]

crimes_gdf = gpd.GeoDataFrame(
    Crime,
    geometry=geometry,
    crs="EPSG:4326"  
)

tracts = tracts.to_crs(crimes_gdf.crs)

crimes_with_tracts = gpd.sjoin(
    crimes_gdf,
    tracts,
    how="left",
    predicate="within"
)

crimes_with_tracts
```

```python
Crime = crimes_with_tracts.iloc[:, :-14]
Crime["Date"] = pd.to_datetime(Crime["Date"])

Battery = Crime[Crime['Primary Type'] == 'BATTERY']
Assault = Crime[Crime['Primary Type'] == 'ASSAULT']
SA = Crime[Crime['Primary Type'].isin(["CRIMINAL SEXUAL ASSAULT", "CRIM SEXUAL ASSAULT"])]
Homicide = Crime[Crime['Primary Type'] == 'HOMICIDE']
DV = Crime[Crime['Primary Type'] == 'DOMESTIC VIOLENCE']

Assault
```

```python
import altair as alt
alt.data_transformers.disable_max_rows()
monthly = (
    Battery
    .groupby(pd.Grouper(key="Date", freq="M"))
    .size()
    .reset_index(name="monthly_count")
)

chart = (
    alt.Chart(monthly)
    .mark_line(point=True)
    .encode(
        x=alt.X(
            "Date:T",
            title="Month"
        ),
        y=alt.Y(
            "monthly_count:Q",
            title="Number of Battery Crimes"
        ),
        tooltip=[
            alt.Tooltip("Date:T", title="Month"),
            alt.Tooltip("monthly_count:Q", title="Count")
        ]
    )
    .properties(
        title="Reported Battery Crimes by Month (2021–2025)",
        width=700,
        height=400
    )
)

chart
```

```python
Assault["Date"] = pd.to_datetime(Assault["Date"])

if "quintile" in Assault.columns:
    Assault = Assault.drop(columns=["quintile"])

tract_counts = (
    Assault
    .groupby("CENSUS_T_1")
    .size()
    .reset_index(name="total_assaults")
)

tract_counts["quintile"] = pd.qcut(
    tract_counts["total_assaults"],
    5,
    labels=["Q1 (Lowest)", "Q2", "Q3", "Q4", "Q5 (Highest)"]
)

Assault = Assault.merge(
    tract_counts[["CENSUS_T_1", "quintile"]],
    on="CENSUS_T_1",
    how="left"
)

Assault["month"] = Assault["Date"].dt.to_period("M").dt.to_timestamp()

quintile_time = (
    Assault
    .groupby(["month", "quintile"])
    .size()
    .reset_index(name="assault_count")
)

chart = (
    alt.Chart(quintile_time)
    .mark_line()
    .encode(
        x=alt.X("month:T", title="Month"),
        y=alt.Y("assault_count:Q", title="Number of Assaults"),
        color=alt.Color("quintile:N", title="Census Tract Quintile"),
    )
    .properties(
        width=700,
        height=400,
        title="Monthly Assaults by Census Tract Quintile"
    )
)

chart
```
